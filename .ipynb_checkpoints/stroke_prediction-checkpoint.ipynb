# Life Style based Stroke Prediction
   
We want to calculate the probability of having a stroke, considering the factors that describe a person's your life style.
We want to find out if the following factors directly/indirectly influence the probability of ever having a stroke, specifically:

- Patient has ever been married
- Patient has ever smoked 
- Which type of work patient does

We will use a publicly available dataset on Kaggle that contains relevant informations about life style and medical conditions of real patients.

### Libraries

Our library of choice is PgmPy, for building a Bayesian Network based on the available data and to do inference on it. We will also use Pandas and Matplotlib to handle the dataset and pictures

```python
import pandas as pd
from pgmpy.inference import VariableElimination
from pgmpy.models import BayesianNetwork
from pgmpy.sampling import BayesianModelSampling
from pgmpy.factors.discrete import State
from pgmpy.base import DAG
from pgmpy.factors.discrete.CPD import TabularCPD
import matplotlib.pyplot as plt
import numpy as np
```

### The Dataset

This dataset is used to predict whether a patient is likely to get stroke based on the input parameters like gender, age, various diseases, and smoking status. It has over 5000 records. Each row provides relavant information about real patients.
We believe the work on this data to be quite inspiring, since Artificial Intelligence finds a great number of applications in the medical field, just like was also shown in class by multiple researchers.
The link to the Kaggle page of the dataset is the following: https://www.kaggle.com/datasets/fedesoriano/stroke-prediction-dataset
_Autore?_

The dataset is in the form of a csv file. We load it in pandas and drop all rows containing at least one NA value, for simplicity's sake. Following is a sample of data and a complete list of the dataset fields.

```python
df = pd.read_csv('healthcare-dataset-stroke-data.csv',sep=',')
display(df.head())

print("Records number, including NAs:",len(df))
df = df.dropna(axis = 0).drop(['id'],axis=1)
print("Records number, after removing NAs:",len(df))
```

We consider the number of lost records due to the NA pruning relatively irrelevant.

# Preprocessing

### 'Unknown' Smoking Status
By analyzing the unique values of the columns, we notice that 'smoking_status' is problematic: it has a significant number of values 'Unknown'. This is equivalent to NA, so we should get rid of them, since we consider the smoking status to be quite relevant.

```python
print('Unique values for the smoking_status field:',df['smoking_status'].unique())
print('Number of records, including Unknown:',len(df))
df = df[df['smoking_status'] != 'Unknown']
print('Number of records, after removing Unknown:',len(df))
```

Notice that we lose a significant amount of data. However it is acceptable, because we are particularly intersted in the smoking status of the patient.

### From Continuous to Categorical

The dataset has a number of continuous values that we wish to make categorical for simplicity's sake.

_Non Ã¨ molto spiegata...?_

```python
CAT_NUM = 4

print('Average glucose level values:',df['avg_glucose_level'].unique())
df['avg_glucose_level'] = pd.cut(df['avg_glucose_level'], CAT_NUM, labels = range(CAT_NUM))
print('Categorical glucose level values:',df['avg_glucose_level'].unique())

# Same discretization goes for bmi and age

df['bmi'] = pd.cut(df['bmi'], CAT_NUM, labels = range(CAT_NUM))
df['age'] = pd.cut(df['age'], CAT_NUM, labels = range(CAT_NUM))

```

# The Network

```python
img = plt.imread('network.png')
plt.figure(figsize = (8,8))
plt.imshow(img)
```

```python
network = [('age', 'ever_married'),
            ('age', 'heart_disease'),
            ('age', 'bmi'),
            ('gender', 'heart_disease'),
            ('gender', 'bmi'),
            ('Residence_type', 'bmi'),
            ('bmi', 'ever_married'),
            ('bmi', 'avg_glucose_level'),
            ('work_type', 'ever_married'),
            ('work_type', 'Residence_type'),
            ('avg_glucose_level', 'hypertension'),
            ('ever_married', 'smoking_status'),
            ('smoking_status', 'stroke'),
            ('smoking_status', 'heart_disease'),
            ('hypertension', 'stroke'),
            ('hypertension', 'heart_disease'),
            ('heart_disease', 'stroke'),
            ]
model = BayesianNetwork(network)
dag = DAG(network)

model.fit(df)
```

### CPD - Smoking Status

PgmPy can calculate CPDs of the different nodes in the graph. It may be interesting to inspect some of them to determine how variables condition each other, or to validate an hypotesis one could make on the data distribution.
Here, for visualization purposes, the default CPD format of PgmPy has been modified.

```python
# build conditional probabilities table

def pretty_cpd(variable,evidence,marginalize,prefix=''):
    print('---------------------------------------------------')
    print('|Marginalized CPD for',variable,'given',evidence)
    print('---------------------------------------------------')
    cpd = model.get_cpds(variable)
    cpd = cpd.marginalize(marginalize,inplace=False)
    print('|',variable.ljust(20),'\t',cpd.state_names[variable][0],'\t',cpd.state_names[variable][1])
    print('---------------------------------------------------')
    for i in range(0,len(cpd.state_names[evidence])):
        print((prefix+str(cpd.state_names[evidence][i])).ljust(20),'\t',round(cpd.values[0][i],2),'\t',round(cpd.values[1][i],2))
    print('\n')  

pretty_cpd('stroke','smoking_status',['hypertension','heart_disease'])
pretty_cpd('stroke','heart_disease',['hypertension','smoking_status'],'Heart Dis.')
pretty_cpd('stroke','hypertension',['smoking_status','heart_disease'],'Hyp.Tens.')
```

Here we can see that, just as one would expect, the presence of a heart disease or hypertension increases the probability of having a stroke. Also smokers tend to have a slightly higher probability of being hit by a stroke.

### Markov Blankets

"In statistics and machine learning, when one wants to infer a random variable with a set of variables, usually a subset is enough, and other variables are useless. Such a subset that contains all the useful information is called a Markov blanket."

Suppose we want to to find out the minimum number of information a patient has to input, in order to have a prediction of the probability of having a stroke. We can first get the Markov Blanket of the stroke node:

```python
model.get_markov_blanket('stroke')
```

The result is quite straight forward, since these are all the nodes right before the stroke node. Smoking Status and Hyeprtension are a quite simple information for the patient to retrieve, however it is difficult to know whether you suffer from heart disease or not. We may wish to relax the problem, and instead of asking the patient if he has heart disease or not, we could ask for the Markov Blanket of Heart Disease:

```python
model.get_markov_blanket('heart_disease')
```

Notice that by asking the patient for:

- age
- gender
- hypertension
- smoking status

we now have the minimum set of nodes for which having any more information (except for heart disease, which we excluded), is not going to of any use. We provide a proof of this reasoning by using d-separation and active trails:

```python
print(dag.active_trail_nodes('stroke',observed=['heart_disease','hypertension','smoking_status']))
print(dag.active_trail_nodes('stroke',observed=['age','gender','hypertension','smoking_status']))
```

# Inference

We are now going to determine if the life style factors are actually relevant to the prediciton of a stroke. To do this, we do inference on the network. Here we use different inference methods at our disposal in PgmPy.

### Exact Inference - Ever Married

Variable Elimination is an exact inference algorithm. This is going to output the exact probability of our query.

```python
# build exact inference 
inference = VariableElimination(model)

print('Evidence: patient has been married')
query = inference.query(['stroke'],{'ever_married':'Yes'},show_progress = False)
print(query)
print('\n')

print('Evidence: patient has never been married')
query = inference.query(['stroke'],{'ever_married':'No'},show_progress = False)
print(query)
print('\n')
```

According to out model, having been married actually increases the probability of hainvg a stroke.

### Rejection Sampling - Work Type

Rejection Sampling is an approximate inference algorithm that samples all nodes and rejects samples that do not respect evidence. This is going to output an approximation of the query probability. The accuracy will increase with the number of samples.

_Usare sample piÃ¹ grande_

```python
import warnings
warnings.filterwarnings('ignore')

approximate_inference = BayesianModelSampling(model)

sample_size = 100
work_types = ['Govt_job','Never_worked','Private','Self-employed','children']

for work_type in work_types:
    query = approximate_inference.rejection_sample(
            evidence=[State('work_type',work_type)], size=sample_size, show_progress=True)
    print('P(stroke|work_type = '+work_type+')',len(query['stroke'][query['stroke']==1])/sample_size)
```

According to out model, work type does not influence probability of stroke.

### Likelihood Weighting - Patient Diagnosis

Likelihood Weighting is an approximate inference algorithm that only samples the nodes not in the evidence. This is going to output an approximation of the query probability. The accuracy will increase with the number of samples.

_Mettere come evidenza le variabili che abbiamo determinato essere decisive con i markov blankets. Potenzialmente dimostrare che l'idea Ã¨ giusta. Forse meglio farlo con exact inference._

```python
query = approximate_inference.likelihood_weighted_sample(
        evidence=[State('age',1),\
                  State('gender','Male'),\
                  State('hypertension',0),\
                  State('smoking_status','never smoked')],\
        size=10000, show_progress=False)

print('Prob is',len(query['stroke'][query['stroke']==1])/100000)

query = approximate_inference.likelihood_weighted_sample(
        evidence=[State('age',1),\
                  State('work_type','Never_worked'),\
                  State('gender','Male'),\
                  State('hypertension',0),\
                  State('smoking_status','never smoked')],\
        size=10000, show_progress=False)

print('Prob is',len(query['stroke'][query['stroke']==1])/100000)
```
